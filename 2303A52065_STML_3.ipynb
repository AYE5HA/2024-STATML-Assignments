```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('/content/sample_data/california_housing_train.csv')
print(d.head())
print(d.tail())
```

```python
mean = d['longitude'].mean()
var = d['longitude'].var()
x = d['longitude']
mle = (1/(np.sqrt(2*np.pi*var)))*np.exp(-((x-mean)**2)/(2*var))
print("Maximum likelihood estimation is", mle)
```

```python
mean = d['latitude'].mean()
var = d['latitude'].var()
def loghd(x):
  return np.log((1/(np.sqrt(2*np.pi*var)))*np.exp(-((x-mean)**2)/(2*var)))
print("Log-likelihood for 50 is:", loghd(50))
print("Log-likelihood for 75 is:", loghd(75))
print("Log-likelihood for 80 is:", loghd(80))
```
